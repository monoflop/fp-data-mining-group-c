```python
#Abhängigkeiten
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
#Daten laden
#strip_1 enthält keine near Einträge mit 1.0?
strip_train = pd.read_csv('data/train/strip_2_train.csv', sep=',')
strip_test = pd.read_csv('data/test/strip_2_test_no_labels.csv', sep=',')

strip_train.head(32)
```

```python
#Label aus Datensatz entfernen, damit wir einen Label-Satz und einen Feature-Satz erhalten.
X_train = strip_train.drop('near',axis = 1)
Y_train = strip_train['near']

#Alle nicht Sensordaten entfernen
X_train = X_train.drop('frame_number',axis = 1)
X_train = X_train.drop('strip_id',axis = 1)
X_train = X_train.drop('node_id',axis = 1)
X_train = X_train.drop('timestamp',axis = 1)
X_train = X_train.drop('run_number',axis = 1)
X_train = X_train.drop('vicon_x',axis = 1)
X_train = X_train.drop('vicon_y',axis = 1)

#NaN Werte normalisieren
X_train = X_train.fillna(X_train.mean())

#X_train.head()

#arr = np.array([], ndmin=2, dtype=complex)
#arr = np.arange(511650, dtype=np.float64).reshape(3411,150)
arr = np.zeros((2229,150), dtype=np.float64)
frame_number = 0
node_index = 0
for i, row in X_train.iterrows():
    #print(str(frame_number))
    arr[frame_number][node_index * 10 + 0] = row['ax']
    arr[frame_number][node_index * 10 + 1] = row['ay']
    arr[frame_number][node_index * 10 + 2] = row['az']
    arr[frame_number][node_index * 10 + 3] = row['gx']
    arr[frame_number][node_index * 10 + 4] = row['gy']
    arr[frame_number][node_index * 10 + 5] = row['gz']
    arr[frame_number][node_index * 10 + 6] = row['mx']
    arr[frame_number][node_index * 10 + 7] = row['my']
    arr[frame_number][node_index * 10 + 8] = row['mz']
    arr[frame_number][node_index * 10 + 9] = row['r']
    node_index = node_index + 1
    if(node_index >= 15):
        frame_number = frame_number + 1
        node_index = 0
        
#TODO error index 2229 is out of bounds for axis 0 with size 2229
        
for index in range(0, 150):
    print(str(index) + " " + str(arr[0][index]))
```

```python
#Daten einheitlich skalieren von z. B. 0.0 - 1.0
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)

sc.fit(X_test)
X_test = sc.transform(X_test)
```

```python
#Random Forest
forest = RandomForestClassifier(n_estimators=200, random_state = 0)
forest.fit(X_train, Y_train)
```

```python
forest_prediction = forest.predict(X_test)

f = open("data.csv", "w")
f.write("Id,Predicted\n")
count = 0
for a in forest_prediction:
    f.write(str(count))
    f.write(",")
    f.write(str(int(a)))
    f.write("\n")
    count = count + 1 
f.close()

#print(forest_prediction)
#print(classification_report(Y_test, forest_prediction))
#print(confusion_matrix(Y_test, forest_prediction))
#print("\nAccuracy = " + str(accuracy_score(Y_test, forest_prediction)))
```
